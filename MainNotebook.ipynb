```python
from DataFetching.audio import GetAudioFromYoutube
audio = GetAudioFromYoutube("https://www.youtube.com/watch?v=9hqWUmyLULo",title="Satu_Tahun_Jokowi",download=False)
audio.list_subtitles()
```

```python
audio.write_subtitle()
```

```python
from DataFetching.vtt2text import convert_vtt
text = convert_vtt("{}.id.vtt".format(audio.title))
text
```

```python
from DataFetching.text_preprocess import *
cleaned_text = text_clean(text)
stemmed_text = text_clean(text_stem(cleaned_text))
shorted_text = remove_short_words(stemmed_text)
word_tokens = nltk.tokenize.word_tokenize(shorted_text)
word_tokens
```

```python
from generate_network import *

edges = generate_edges(word_tokens)
G = create_graph_from_edges(edges, with_tsne=True)
color_connected_components(G)

net = Network(notebook=True,heading=audio.title)
net.width = '100%'
net.from_nx(G, edge_weight_transf=float)
net.set_options("""
var options = {
"nodes": {
"font": {
  "size": 40
}
}
}
""")
net.show("word.html")
```
